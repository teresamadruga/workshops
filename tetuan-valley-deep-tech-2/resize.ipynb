# Resize image
Simple notebook that shows how to resize an image.

```python
from PIL import Image
from os.path import exists
```

```python
def resize_image(input_image_path, output_image_path, new_size):
    image = Image.open(input_image_path)
    resized_image = image.resize(new_size)
    resized_image.save(output_image_path)
```

### Example of usage
Replace `input_path` with your input image path, replace `output_path` with your desired output image path, replace `new_size` with your desired new size (width, height), and run the `resize_image` function. The example has a check to not override an existing file. 

`woman-cat-*.jpg` image source: Picture from <a href="https://unsplash.com/@microsoft365?utm_source=unsplash&utm_medium=referral&utm_content=creditCopyText">Microsoft 365</a> in <a href="https://unsplash.com/es/fotos/f1zQuagWCTA?utm_source=unsplash&utm_medium=referral&utm_content=creditCopyText">Unsplash</a>

```python
input_path = "images/woman-cat-square.jpg" 
output_path = "images/woman-cat-512.jpg"
new_size = (512, 512)



file_exists = exists(output_path)

if file_exists:
    print("The output path already exists")
else:
    resize_image(input_path, output_path, new_size)
```
